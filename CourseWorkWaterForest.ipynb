```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#  
alpha_A = 0.1
beta_A = 0.05
gamma_A = 0.03
delta_A = 0.02

alpha_B = 0.07
beta_B = 0.04
gamma_B = 0.06
delta_B = 0.05

#  
W0 = 100
L0 = 50
E_A0_star = 10
E_B0_star = 10

#    r_A  r_B
k_A = 0.01
m_A = 0.02
k_B = 0.02
m_B = 0.01

#  r_A  r_B
def r_A(W, L):
    return k_A * W + m_A * L

def r_B(W, L):
    return k_B * W + m_B * L

#      
E_A0 = E_A0_star * r_A(W0, L0)
E_B0 = E_B0_star * r_B(W0, L0)

#  
def model(y, t):
    W, L, E_A, E_B = y
    dWdt = -alpha_A * E_A + delta_A * W - beta_B * L
    dLdt = -alpha_B * E_B
    dE_Adt = beta_A * W - gamma_A * E_A
    dE_Bdt = gamma_B * W - delta_B * E_B
    return [dWdt, dLdt, dE_Adt, dE_Bdt]

#  
t = np.linspace(0, 50, 500)

#  
y0 = [W0, L0, E_A0, E_B0]

#   
solution = odeint(model, y0, t)

#   
W = solution[:, 0]
L = solution[:, 1]
E_A = solution[:, 2]
E_B = solution[:, 3]

plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.plot(t, W, 'b')
plt.xlabel('')
plt.ylabel('  (W)')
plt.title('  ')

plt.subplot(2, 2, 2)
plt.plot(t, L, 'g')
plt.xlabel('')
plt.ylabel('  (L)')
plt.title('  ')

plt.subplot(2, 2, 3)
plt.plot(t, E_A, 'r')
plt.xlabel('')
plt.ylabel('  A (E_A)')
plt.title('   A')

plt.subplot(2, 2, 4)
plt.plot(t, E_B, 'k')
plt.xlabel('')
plt.ylabel('  B (E_B)')
plt.title('   B')

plt.tight_layout()
plt.show()
```

```python
#   
import numpy as np
import matplotlib.pyplot as plt

#   ( ,     )
alpha_1 = 0.1
alpha_2 = 0.05
k_1 = 0.03
k_2 = 0.04

#  
V1_0 = 50
L2_0 = 50

#  
num_steps = 20
delta_t = 0.01  #      
t = np.linspace(0, num_steps * delta_t, num_steps + 1)

#    
V1 = np.zeros(num_steps + 1)
L2 = np.zeros(num_steps + 1)

#  
V1[0] = V1_0
L2[0] = L2_0

#   RK4
def dV1_dt(V1, alpha_1, k_1, L2):
    V2 = (alpha_1 / k_1) * V1
    return alpha_1 * V2 * L2 - k_1 * V1

def dL2_dt(V1, alpha_1, alpha_2, k_1, k_2, L2):
    V2 = (alpha_1 / k_1) * V1
    return alpha_2 * V2 * L2 - k_2 * L2

#    
def strategy_A(step):
    #  A   alpha_1  k_1
    if step % 1 == 0:  #    10 
        return np.random.uniform(-0.1, 0.1), np.random.uniform(-0.03, 0.03)
    return alpha_1, k_1

def strategy_B(step):
    #  B   alpha_2  k_2
    if step % 1 == 0:  #    10 
        return np.random.uniform(-0.05, 0.05), np.random.uniform(-0.04, 0.04)
    return alpha_2, k_2

#   RK4    
for n in range(num_steps):
    alpha_1, k_1 = strategy_A(n)
    alpha_2, k_2 = strategy_B(n)
    
    #    V1
    k1_V1 = delta_t * dV1_dt(V1[n], alpha_1, k_1, L2[n])
    k2_V1 = delta_t * dV1_dt(V1[n] + 0.5 * k1_V1, alpha_1, k_1, L2[n] + 0.5 * k1_V1)
    k3_V1 = delta_t * dV1_dt(V1[n] + 0.5 * k2_V1, alpha_1, k_1, L2[n] + 0.5 * k2_V1)
    k4_V1 = delta_t * dV1_dt(V1[n] + k3_V1, alpha_1, k_1, L2[n] + k3_V1)

    V1[n+1] = V1[n] + (k1_V1 + 2 * k2_V1 + 2 * k3_V1 + k4_V1) / 6

    #    L2
    k1_L2 = delta_t * dL2_dt(V1[n], alpha_1, alpha_2, k_1, k_2, L2[n])
    k2_L2 = delta_t * dL2_dt(V1[n] + 0.5 * k1_L2, alpha_1, alpha_2, k_1, k_2, L2[n] + 0.5 * k1_L2)
    k3_L2 = delta_t * dL2_dt(V1[n] + 0.5 * k2_L2, alpha_1, alpha_2, k_1, k_2, L2[n] + 0.5 * k2_L2)
    k4_L2 = delta_t * dL2_dt(V1[n] + k3_L2, alpha_1, alpha_2, k_1, k_2, L2[n] + k3_L2)

    L2[n+1] = L2[n] + (k1_L2 + 2 * k2_L2 + 2 * k3_L2 + k4_L2) / 6

#   
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, V1, 'b')
plt.xlabel('')
plt.ylabel('    A (V1)')
plt.title('     A')

plt.subplot(2, 1, 2)
plt.plot(t, L2, 'r')
plt.xlabel('')
plt.ylabel('    B (L2)')
plt.title('     B')

plt.tight_layout()
plt.show()
```
